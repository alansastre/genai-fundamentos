## Configuración inicial

Anthropic requiere una API key para autenticar las peticiones. Obtén tu clave desde [platform.anthropic.com](https://platform.claude.com/settings/keys) en la sección **API Keys**.

Instala el paquete de integración:

```bash .noeval
pip install langchain-anthropic
```

Crea un archivo `.env` con tu clave:

```.noeval
ANTHROPIC_API_KEY=sk-ant-api03-tu-clave-aqui
```

La clase **ChatAnthropic** detecta automáticamente la variable de entorno:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5")

respuesta = modelo.invoke("¿Cuál es la capital de Francia?")
print(respuesta.content)
````

> Añade `.env` a tu `.gitignore` para no exponer credenciales.

## Modelos disponibles

Anthropic organiza sus modelos en tres niveles con nombres inspirados en la música:

| Modelo | Uso recomendado |
|--------|-----------------|
| **claude-haiku-4-5** | Rápido y económico, ideal para desarrollo |
| **claude-sonnet-4-5** | Balance óptimo entre rendimiento y coste |

````python
from langchain_anthropic import ChatAnthropic

# Modelo económico para pruebas
modelo_haiku = ChatAnthropic(model="claude-haiku-4-5")
respuesta = modelo_haiku.invoke("Explica qué es Python en una frase")
print(f"Claude Haiku: {respuesta.content}")
````

````python
from langchain_anthropic import ChatAnthropic

# Modelo con mayor capacidad
modelo_sonnet = ChatAnthropic(model="claude-sonnet-4-5")
respuesta = modelo_sonnet.invoke("Explica qué es Python en una frase")
print(f"Claude Sonnet: {respuesta.content}")
````

## Conversaciones con mensajes

Los modelos de Anthropic soportan los formatos de mensajes de LangChain:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5")

mensajes = [
    ("system", "Eres un experto en geografía. Responde de forma concisa."),
    ("human", "¿Cuál es el río más largo del mundo?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

Claude destaca por su **adherencia estricta** a las instrucciones del system prompt, lo que lo hace ideal para aplicaciones donde el formato de salida es importante.

## Parámetros de generación

Puedes ajustar parámetros como **temperature** para controlar la creatividad:

````python
from langchain_anthropic import ChatAnthropic

# Respuesta más determinista
modelo_preciso = ChatAnthropic(model="claude-haiku-4-5", temperature=0.0)
respuesta = modelo_preciso.invoke("¿Cuánto es 2 + 2?")
print(f"Temperature 0: {respuesta.content}")

# Respuesta más creativa
modelo_creativo = ChatAnthropic(model="claude-haiku-4-5", temperature=1.0)
respuesta = modelo_creativo.invoke("Inventa un nombre para una cafetería")
print(f"Temperature 1: {respuesta.content}")
````

## Monitorizar uso de tokens

Controlar el consumo de tokens ayuda a gestionar costes:

````python
from langchain_anthropic import ChatAnthropic

modelo = ChatAnthropic(model="claude-haiku-4-5")
respuesta = modelo.invoke("Resume en una frase las ventajas de Python")

print(f"Respuesta: {respuesta.content}")
print(f"Tokens entrada: {respuesta.usage_metadata.get('input_tokens')}")
print(f"Tokens salida: {respuesta.usage_metadata.get('output_tokens')}")
print(f"Total: {respuesta.usage_metadata.get('total_tokens')}")
````

## Contar tokens antes de enviar

Una característica útil de Anthropic es poder **estimar tokens** antes de enviar la petición:

````python
from langchain_anthropic import ChatAnthropic
from langchain.messages import HumanMessage, SystemMessage

modelo = ChatAnthropic(model="claude-haiku-4-5")

mensajes = [
    SystemMessage(content="Eres un asistente útil"),
    HumanMessage(content="Explica la teoría de cuerdas en términos muy simples y de manera breve")
]

num_tokens = modelo.get_num_tokens_from_messages(mensajes)
print(f"Tokens estimados en el mensaje: {num_tokens}")
````

Esto es útil para validar que el mensaje no excede el límite del contexto o para estimar costes.
