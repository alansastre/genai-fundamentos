## Configuración inicial

OpenAI requiere una API key para autenticar las peticiones. Obtén tu clave desde [platform.openai.com](https://platform.openai.com) en la sección **API keys**.

Instala el paquete de integración:

```bash .noeval
pip install langchain-openai
```

Crea un archivo `.env` con tu clave:

```.noeval
OPENAI_API_KEY=sk-proj-tu-clave-aqui
```

La clase **ChatOpenAI** detecta automáticamente la variable de entorno:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-nano")

respuesta = modelo.invoke("¿Cuál es la capital de Francia?")
print(respuesta.content)
````

> Añade `.env` a tu `.gitignore` para no exponer credenciales.

## Modelos disponibles

OpenAI ofrece varios modelos con diferentes capacidades y costes:

| Modelo | Uso recomendado |
|--------|-----------------|
| **gpt-5-nano** | Pruebas y desarrollo, muy económico |
| **gpt-5-mini** | Desarrollo, buen balance coste/calidad |
| **gpt-5** | Producción, alta calidad |
| **gpt-5.1** | Tareas complejas |
| **gpt-5.2** | Máxima capacidad, razonamiento avanzado |

````python
from langchain_openai import ChatOpenAI

# Modelo económico para pruebas
modelo_nano = ChatOpenAI(model="gpt-5-nano")
respuesta = modelo_nano.invoke("Explica qué es Python en una frase")
print(f"GPT-5-nano: {respuesta.content}")
````

````python
from langchain_openai import ChatOpenAI

# Modelo con mayor capacidad
modelo_mini = ChatOpenAI(model="gpt-5-mini")
respuesta = modelo_mini.invoke("Explica qué es Python en una frase")
print(f"GPT-5-mini: {respuesta.content}")
````

## Conversaciones con mensajes

Los modelos de OpenAI soportan los formatos de mensajes de LangChain:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-nano")

mensajes = [
    ("system", "Eres un experto en geografía. Responde de forma concisa."),
    ("human", "¿Cuál es el río más largo del mundo?")
]

respuesta = modelo.invoke(mensajes)
print(respuesta.content)
````

El **system prompt** define el comportamiento del modelo. El modelo seguirá estas instrucciones durante toda la conversación.

## Parámetros de generación

Puedes ajustar parámetros como **temperature** para controlar la creatividad:

````python
from langchain_openai import ChatOpenAI

# Respuesta más determinista (útil para tareas precisas)
modelo_preciso = ChatOpenAI(model="gpt-5-nano", temperature=0.0)
respuesta = modelo_preciso.invoke("¿Cuánto es 2 + 2?")
print(f"Temperature 0: {respuesta.content}")

# Respuesta más creativa (útil para generación de contenido)
modelo_creativo = ChatOpenAI(model="gpt-5-nano", temperature=1.0)
respuesta = modelo_creativo.invoke("Inventa un nombre para una cafetería")
print(f"Temperature 1: {respuesta.content}")
````

## Monitorizar uso de tokens

Controlar el consumo de tokens ayuda a gestionar costes:

````python
from langchain_openai import ChatOpenAI

modelo = ChatOpenAI(model="gpt-5-nano")
respuesta = modelo.invoke("Resume en una frase las ventajas de Python")

print(f"Respuesta: {respuesta.content}")
print(f"Tokens entrada: {respuesta.usage_metadata.get('input_tokens')}")
print(f"Tokens salida: {respuesta.usage_metadata.get('output_tokens')}")
print(f"Total: {respuesta.usage_metadata.get('total_tokens')}")
````

> Los tokens consumidos determinan el coste de cada petición. Los modelos nano y mini son significativamente más económicos.
