La IA Generativa no surgió de la nada. Es el resultado de décadas de investigación. Para un desarrollador que inicia, basta con conocer los **momentos clave** que han definido tanto la tecnología como las herramientas que usamos hoy.

## 2012: Deep Learning funciona

Durante décadas, las redes neuronales fueron una idea prometedora pero poco práctica. En 2012, **AlexNet** demostró que las redes profundas podían superar a los métodos tradicionales en tareas de visión por computador.

El cambio fue posible gracias a dos factores:
- **GPUs**: permitieron entrenar redes mucho más grandes
- **Datos masivos**: internet proporcionó millones de ejemplos de entrenamiento

> Este momento marcó el inicio de la era del Deep Learning: las redes neuronales dejaron de ser un tema académico para convertirse en una herramienta práctica.

## 2014-2020: Primera generación de herramientas para código

Mientras el Deep Learning maduraba, surgieron las primeras herramientas que aplicaban **machine learning tradicional** (no LLMs) al código:

| Herramienta | Año | Características |
|-------------|-----|-----------------|
| **Kite** | 2014 | Autocompletado inteligente para Python |
| **Tabnine** | 2015 | Sugerencias basadas en patrones locales |
| **IntelliCode** | 2018 | Extensión de Microsoft para Visual Studio |

Estas herramientas:
- Completaban nombres de variables y funciones
- Sugerían la siguiente línea basándose en patrones estadísticos
- Funcionaban sin conexión, entrenadas con código público

**Limitación**: solo predecían texto inmediato basándose en frecuencias, sin entender semántica ni contexto del proyecto.

> Kite cerró en 2022, pero demostró que había demanda real de asistentes de código con IA.

## 2017: La arquitectura Transformer

El paper ["Attention Is All You Need"](https://arxiv.org/abs/1706.03762) introdujo la arquitectura **Transformer**, que revolucionó el procesamiento de texto. Su innovación clave fue el mecanismo de **atención**, que permite procesar secuencias en paralelo capturando relaciones entre todas las palabras simultáneamente.

El paper original propuso una arquitectura **encoder-decoder** para traducción, pero de ahí surgieron **tres familias** que hoy coexisten:

| Familia | Estructura | Qué hace bien | Ejemplos actuales |
|---------|------------|---------------|-------------------|
| **Encoder-only** | Solo codifica | Comprender y clasificar texto | BERT, RoBERTa, modelos de embeddings |
| **Decoder-only** | Solo genera | Crear texto nuevo | GPT, Claude, Gemini, LLaMA |
| **Encoder-decoder** | Ambos | Transformar texto (traducir, resumir) | T5, BART, modelos de traducción |

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart TB
    T["Transformer original<br/>(2017)"]
    T --> EO["Encoder-only<br/>Comprensión"]
    T --> DO["Decoder-only<br/>Generación"]
    T --> ED["Encoder-decoder<br/>Transformación"]

    style T fill:#9f7aea,color:#fff
```

> Las tres familias siguen usándose hoy. Los **LLM conversacionales** (GPT, Claude) son decoder-only, pero los **modelos de embeddings** para búsqueda semántica suelen ser encoder-only.

## 2018: GPT y BERT

En 2018, Google y OpenAI demostraron el potencial de dos familias diferentes:

| Modelo | Familia | Cómo funciona | Uso principal |
|--------|---------|---------------|---------------|
| **BERT** (Google) | Encoder-only | Mira todo el texto a la vez | Clasificación, embeddings, búsqueda |
| **GPT** (OpenAI) | Decoder-only | Predice la siguiente palabra | Generación de texto |

**BERT** procesa el texto **bidireccionalmente** (ve pasado y futuro), lo que lo hace excelente para entender significado. Por eso los modelos de embeddings actuales derivan de esta familia.

**GPT** es **autoregresivo**: genera texto palabra a palabra, usando solo lo anterior. Este enfoque simple pero potente escaló mejor y se convirtió en la base de los LLM conversacionales actuales.

> No es que decoder-only "ganó". Cada familia resuelve problemas diferentes: GPT genera, BERT comprende. Muchos sistemas de IA combinan ambos tipos.

## 2020: GPT-3 y las leyes de escalado

OpenAI demostró con **GPT-3** que hacer modelos más grandes producía mejoras dramáticas. Las **leyes de escalado** formalizaron esta relación:

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart LR
    P["Más parámetros"] --> E["Mejores capacidades"]
    D["Más datos"] --> E
    C["Más cómputo"] --> E

    style E fill:#9f7aea,color:#fff
```

GPT-3 con 175 mil millones de parámetros mostró **capacidades emergentes**: habilidades que no estaban presentes en modelos más pequeños, como escribir código o razonar sobre problemas.

> La lección de GPT-3 fue clara: escalar funciona. Esto desató una carrera entre empresas por entrenar modelos cada vez más grandes.

## 2021: GitHub Copilot - LLMs para código

**GitHub Copilot** fue el primer producto masivo que llevó los LLMs al desarrollo de software:

- **Preview técnico**: junio de 2021
- **Disponibilidad general**: junio de 2022
- Basado en **Codex**, una versión de GPT-3 especializada en código

Capacidades que revolucionaron el mercado:
- Generar funciones completas a partir de comentarios
- Entender docstrings y crear implementaciones coherentes
- Sugerir código en múltiples lenguajes
- Integrarse directamente en VS Code

**Impacto**: Por primera vez, millones de desarrolladores usaron LLMs en su trabajo diario. Las herramientas de autocompletado estadístico quedaron obsoletas.

**Limitación**: seguía siendo reactivo, solo sugería código línea a línea sin poder ejecutar acciones ni mantener conversaciones.

## Noviembre 2022: ChatGPT democratiza los LLMs

Antes de ChatGPT, los LLMs eran herramientas para investigadores. En noviembre de 2022, OpenAI lanzó **ChatGPT** y todo cambió:

- Interfaz de chat accesible para cualquier persona
- Respuestas útiles y coherentes gracias a **RLHF** (aprendizaje por feedback humano)
- 100 millones de usuarios en 2 meses

El secreto de ChatGPT no fue solo el modelo, sino cómo fue **alineado** con preferencias humanas. En lugar de solo predecir texto, aprendió a dar respuestas que los usuarios encontraban útiles.

> ChatGPT fue el "momento iPhone" de la IA: demostró que la tecnología estaba lista para el uso masivo.

**Impacto en desarrolladores**: Millones empezaron a usar ChatGPT para generar código, depurar errores y aprender tecnologías. Esto catalizó la carrera por integrar chat en los IDEs.

## 2023: Chat integrado en IDEs

La incorporación de **chat dentro del IDE** permitió conversaciones sobre el código sin salir del editor:

| Herramienta | Lanzamiento | Características |
|-------------|-------------|-----------------|
| **Copilot Chat** | 2023 | Chat integrado en VS Code |
| **Cursor** | 2023 | IDE basado en VS Code con IA nativa |
| **Codeium** | 2023 | Alternativa gratuita con chat |

Capacidades nuevas:
- Explicar código existente en lenguaje natural
- Refactorizar selecciones de código mediante diálogo
- Detectar y corregir bugs conversando
- Generar tests a partir del código fuente

**Limitación**: el desarrollador aún debía copiar, pegar y aplicar manualmente muchos cambios sugeridos.

## 2023-2024: Multimodalidad y competencia

Tras ChatGPT, la industria explotó con nuevos modelos y capacidades:

- **GPT-4** (marzo 2023): primer modelo multimodal de OpenAI (texto + imagen)
- **Claude** (Anthropic): enfocado en seguridad y contextos largos
- **Gemini** (Google): multimodal nativo desde el inicio
- **LLaMA** (Meta): modelos open-weight que democratizaron el acceso

La **multimodalidad** se convirtió en estándar: los modelos actuales pueden procesar y generar texto, imágenes, audio y video de forma integrada.

## 2024: Agentes en IDE y herramientas generativas

En 2024 surgieron dos tendencias paralelas que transformaron el desarrollo:

### Agentes en IDE

Herramientas que no solo sugieren, sino que **ejecutan cambios** directamente:

- **Cursor Agent mode**: aplica cambios en múltiples archivos automáticamente
- **Windsurf** (antes Codeium): IDE con capacidades agénticas completas

### Herramientas generativas web

Plataformas que crean aplicaciones completas desde descripciones:

| Herramienta | Fecha | Qué hace |
|-------------|-------|----------|
| **Devin** | Marzo 2024 | Primer "ingeniero de software IA" autónomo |
| **v0** (Vercel) | 2024 | Genera interfaces React desde texto |
| **Bolt.new** | 2024 | Crea aplicaciones web funcionales |
| **Lovable** | 2024 | Genera apps completas con backend |
| **Copilot Workspace** | Abril 2024 | Desarrollo colaborativo con IA en GitHub |

## 2025: Vibe Coding y CLIs inteligentes

Andrej Karpathy acuñó el término **Vibe Coding** en febrero de 2025 para describir la nueva forma de programar:

- El desarrollador **describe la intención** en lenguaje natural
- El agente interpreta, genera el código y aplica los cambios automáticamente
- El desarrollador revisa, ajusta y guía mediante conversación
- El ciclo se repite hasta conseguir el resultado deseado

El término fue elegido **palabra del año 2025**, reflejando su adopción masiva.

> Linus Torvalds expresó que el Vibe Coding es útil para aprendizaje y proyectos personales, pero advirtió contra su uso en sistemas críticos sin supervisión experta.

Otra tendencia importante de 2025 ha sido el auge de los **agentes CLI** (Command Line Interface):

| Herramienta | Lanzamiento | Características |
|-------------|-------------|-----------------|
| **Claude Code** | Febrero 2025 | Agente de Anthropic, comprende todo el proyecto |
| **OpenAI Codex CLI** | 2025 | Agente oficial de OpenAI para terminal |
| **Cursor CLI** | 2025 | Extensión del IDE Cursor para la terminal |
| **Gemini CLI** | 2025 | Agente de Google integrado en Google Cloud SDK |
| **OpenCode** | 2025 | Agente open-source agnóstico de modelo |

**Claude Code** se ha convertido en uno de los más populares porque:

- Opera directamente desde la terminal, sin necesidad de IDE específico
- Comprende el proyecto completo y su contexto
- Puede editar archivos, ejecutar comandos y gestionar Git
- Mantiene transparencia total sobre las acciones que realiza
- Se integra en cualquier flujo de trabajo existente

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart LR
    T["Terminal"] --> CC["Claude Code"]
    CC --> A1["Analiza codebase"]
    CC --> A2["Edita archivos"]
    CC --> A3["Ejecuta tests"]
    CC --> A4["Crea commits"]

    style CC fill:#7c4dff,color:#fff
```

> Los agentes CLI representan un cambio filosófico: en lugar de adaptar el flujo de trabajo al IDE, el agente se adapta al flujo del desarrollador.
