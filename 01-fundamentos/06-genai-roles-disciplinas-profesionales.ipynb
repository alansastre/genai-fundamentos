## El panorama profesional actual

La IA Generativa ha creado nuevas disciplinas profesionales. A diferencia del ML tradicional, estos roles se centran en **usar modelos existentes** para construir aplicaciones, no en entrenar modelos desde cero.

La inversión global en IA crece cada año, con demanda de profesionales en todo el mundo: EEUU, Europa, Asia-Pacífico y mercados emergentes.

> Estos roles no requieren entrenar modelos. La habilidad está en saber configurarlos, integrarlos y operarlos correctamente.

## Prompt Engineering: skill transversal

El **Prompt Engineering** no es un puesto de trabajo independiente. Es una **skill transversal** que todos los profesionales de IA necesitan dominar.

### Por qué no es un rol separado

- Los modelos mejoran constantemente en interpretar instrucciones
- Las empresas entrenan a empleados existentes en lugar de contratar especialistas
- Las técnicas de prompting se integran en herramientas y frameworks
- El valor está en combinar prompting con conocimiento de dominio

### Técnicas fundamentales que debes conocer

| Técnica | Descripción | Cuándo usarla |
|---------|-------------|---------------|
| **Zero-shot** | Pedir la tarea sin ejemplos | Tareas simples y claras |
| **Few-shot** | Dar ejemplos antes de la tarea | Cuando el formato importa |
| **Chain of Thought** | Pedir razonamiento paso a paso | Problemas complejos |
| **Role prompting** | Asignar un rol al modelo | Tareas especializadas |

### Context Engineering

El prompting ha evolucionado hacia **Context Engineering**: gestionar todo lo que el modelo recibe.

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart TB
    subgraph contexto["Contexto completo"]
        SYS["System prompt"]
        TOOLS["Herramientas"]
        RAG["Datos externos"]
        HIST["Historial"]
        MEM["Memoria"]
    end

    contexto --> MOD["Modelo"]
    MOD --> RES["Respuesta"]

    style contexto fill:#9f7aea,color:#fff
```

El contexto incluye:
- **System prompt**: las instrucciones base
- **Herramientas**: funciones que el modelo puede invocar
- **Datos externos**: información recuperada (RAG)
- **Historial**: conversaciones previas
- **Memoria**: estado persistente entre sesiones

> El arte del Context Engineering está en incluir la información justa: ni demasiada (satura el modelo) ni poca (falta contexto crítico).

## AI Engineer

El **AI Engineer** es el rol central en el ecosistema de IA Generativa. Construye aplicaciones completas que integran modelos de IA. No entrena modelos, los **orquesta**.

### Diferencia con ML Engineer

| ML Engineer | AI Engineer |
|-------------|-------------|
| Entrena y optimiza modelos | Integra modelos en aplicaciones |
| Prepara datasets de entrenamiento | Gestiona contexto y prompts |
| Output: modelo entrenado | Output: aplicación funcional |
| Focus: métricas del modelo | Focus: experiencia del usuario |

### Responsabilidades típicas

Un AI Engineer:
- Diseña la **arquitectura** de la aplicación
- Implementa pipelines de **RAG** (Retrieval Augmented Generation)
- Gestiona el **contexto** en conversaciones
- Integra **herramientas** externas y APIs
- Optimiza **latencia** y **costes**
- Asegura **calidad** de las respuestas

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart TB
    AIE["AI Engineer"] --> APP["Aplicación"]

    LLM["APIs LLM"] --> APP
    VDB["Vector DB"] --> APP
    TOOLS["Herramientas"] --> APP
    MON["Monitorización"] --> APP

    style AIE fill:#9f7aea,color:#fff
```

### Skills requeridas

| Área | Habilidades |
|------|-------------|
| **Programación** | Cualquier lenguaje backend (existen frameworks de IA para todos los stacks) |
| **APIs de LLMs** | Integración con proveedores, streaming, manejo de errores |
| **Orquestación** | Frameworks de orquestación de LLMs del ecosistema elegido |
| **Datos** | Bases de datos vectoriales, RAG, embeddings |
| **Infraestructura** | Contenedores, cloud, despliegue de aplicaciones |
| **Producto** | UX de IA, feedback loops, gestión de expectativas |

> El AI Engineering es **agnóstico de lenguaje**. Existen frameworks maduros para Python, Java, TypeScript y otros. Lo importante son los conceptos, no la herramienta específica.

## Agentic AI Developer

El **Agentic AI Developer** (también llamado Agent Engineer) diseña agentes autónomos capaces de ejecutar tareas complejas con mínima supervisión.

### Qué diferencia a un agente

Un agente es un sistema donde un **LLM controla un bucle de ejecución**. En cada iteración decide si necesita invocar herramientas o si ya puede responder al usuario.

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart TB
    INPUT["Tarea del usuario"] --> LLM["LLM<br/>(razona)"]
    
    LLM --> DEC{{"¿Necesita<br/>herramienta?"}}
    
    DEC -->|Sí| TOOL["Ejecutar<br/>herramienta"]
    TOOL --> OBS["Observación<br/>(resultado)"]
    OBS --> LLM
    
    DEC -->|No| RESP["Respuesta<br/>al usuario"]

    style LLM fill:#9f7aea,color:#fff
    style DEC fill:#fef3c7
```

Este **tool loop** es la diferencia fundamental:
- **Sin agente**: el LLM responde directamente, sin acceso a información externa
- **Con agente**: el LLM puede invocar herramientas, observar resultados, y seguir razonando hasta completar la tarea

### Responsabilidades específicas

Un Agentic AI Developer:
- Diseña qué **herramientas** tiene el agente
- Define **guardrails**: límites de lo que puede hacer
- Implementa puntos de **supervisión humana** (human-in-the-loop)
- Gestiona **estado y memoria** entre interacciones
- Orquesta **múltiples agentes** cuando es necesario
- Asegura **observabilidad** y logging de acciones

### Técnicas y patrones clave

| Patrón | Descripción |
|--------|-------------|
| **ReAct** | Razonamiento y acción intercalados |
| **Tool calling** | Invocación estructurada de funciones externas |
| **Multi-agent** | Coordinación entre agentes especializados |
| **Workflows con estado** | Grafos de ejecución con persistencia |
| **Human-in-the-loop** | Puntos de aprobación humana en flujos críticos |

> Existen frameworks de agentes para todos los lenguajes principales. El Agentic AI Development requiere pensar en sistemas que operan con incertidumbre: es diseño de sistemas + comprensión profunda de LLMs.

## LLMOps Engineer

**LLMOps** se centra en llevar aplicaciones de IA Generativa a producción de forma fiable y escalable.

### Diferencia con MLOps

| Aspecto | MLOps | LLMOps |
|---------|-------|--------|
| Entrenamiento | Frecuente | Poco frecuente |
| Evaluación | Métricas numéricas | Evaluación cualitativa |
| Costes principales | Entrenamiento | Inferencia |
| Versionado | Pesos del modelo | Prompts y configuraciones |
| Latencia | Menos crítica | Muy crítica (UX) |

### Áreas de responsabilidad

Un LLMOps Engineer gestiona:
- **Despliegue**: infraestructura con baja latencia
- **Monitorización**: detectar degradación, drift y alucinaciones
- **Costes**: optimizar tokens, implementar caching, routing de modelos
- **Versionado**: control de cambios en prompts y configuraciones
- **Evaluación**: testing continuo de calidad

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart LR
    DEP["Despliegue"] --> PROD["Producción fiable"]
    MON["Monitorización"] --> PROD
    COST["Costes"] --> PROD
    EVAL["Evaluación"] --> PROD
    VER["Versionado"] --> PROD

    style PROD fill:#9f7aea,color:#fff
```

### Capacidades del ecosistema LLMOps

| Categoría | Qué resuelve |
|-----------|--------------|
| **Observabilidad** | Tracing de llamadas, latencia, tokens consumidos |
| **Evaluación** | Testing automatizado de calidad de respuestas |
| **Gateway/Routing** | Abstracción de proveedores, balanceo, fallbacks |
| **Caching** | Reutilización de respuestas similares, reducción de costes |
| **Guardrails** | Validación de inputs/outputs, filtrado de contenido |

> LLMOps es esencial cuando pasas de prototipos a sistemas con usuarios reales. El ecosistema de herramientas evoluciona rápido, pero las capacidades que necesitas son estables.

## Roles emergentes

### AI Ethics & Governance

Rol enfocado en **cumplimiento normativo** y uso responsable de IA:
- Auditoría de sesgos en modelos y datos
- Cumplimiento de regulaciones (EU AI Act, GDPR)
- Políticas de uso de IA en la organización
- Evaluación de impacto ético

### AI Security Engineer

Especialista en **seguridad de sistemas de IA**:
- Protección contra prompt injection
- Defensa contra ataques adversariales
- Auditoría de vulnerabilidades
- Data leakage prevention

### AI Trust & Safety

Rol que asegura que los sistemas de IA sean **confiables**:
- Detección de contenido problemático
- Sistemas de moderación
- Evaluación de fiabilidad
- Red teaming de modelos

> Estos roles emergentes tienen **escasez crítica de talento**. Son altamente demandados pero pocos profesionales están cualificados.

## Roles que crean los modelos

Si estos roles se centran en **usar** modelos existentes, ¿quién los crea? Aquí entran roles de ML y Deep Learning tradicionales.

| Perfil | Enfoque principal | Cuándo crea modelos |
|--------|-------------------|---------------------|
| **Data Scientist** | Análisis, experimentación, métricas | Modelos para casos de negocio acotados |
| **ML Engineer** | Entrenamiento y despliegue | Modelos para producción (no LLMs fundacionales) |
| **Research Scientist** | Investigación de arquitecturas | Define cómo debe entrenarse un modelo nuevo |
| **Pretraining Engineer** | Entrenamiento a gran escala | LLMs fundacionales en infraestructura real |

> "Crear un LLM fundacional desde cero" es trabajo de equipos de **investigación** e **ingeniería de preentrenamiento** en empresas como OpenAI, Anthropic, Google o Meta.
