## Qué es la inteligencia artificial

La **Inteligencia Artificial** (IA) es el campo de la informática dedicado a crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana. Estas tareas incluyen reconocer imágenes, comprender lenguaje natural, tomar decisiones y resolver problemas.

A diferencia de la programación tradicional, donde el desarrollador escribe reglas explícitas para cada situación, los sistemas de IA pueden **aprender patrones** y adaptarse a nuevos escenarios.

```mermaid
%%{init: {'theme': 'default'}}%%

    end

    subgraph ia["Inteligencia Artificial"]

    end

```

Un programa tradicional para detectar spam requeriría que alguien definiera manualmente todas las palabras sospechosas. Un sistema de IA aprende a identificar spam analizando miles de ejemplos.

> La IA no es una tecnología única, sino un **campo amplio** que engloba múltiples enfoques y técnicas para simular capacidades cognitivas.

## Machine Learning: aprendizaje automático

El **Machine Learning** (ML) o aprendizaje automático permite a los sistemas aprender de los datos sin ser programados explícitamente para cada tarea. En lugar de escribir instrucciones detalladas, se entrena un modelo con ejemplos.

El proceso de Machine Learning sigue un patrón común:

```mermaid
%%{init: {'theme': 'default'}}%%

        direction TB
        D1["Datos<br/>etiquetados"] --> A1["Algoritmo"]
        A1 --> M1["Modelo<br/>entrenado"]
    end

    subgraph infer["Inferencia"]
        direction TB
        D2["Dato<br/>nuevo"] --> M2["Modelo"]
        M2 --> P["Predicción"]
    end

    M1 -.->|"se usa"| M2

    style train fill:#e8f5e9,stroke:#2e7d32
    style infer fill:#fff3e0,stroke:#ef6c00
```

> El Machine Learning transforma la forma de resolver problemas: en lugar de programar *cómo* resolver algo, proporcionamos *ejemplos* de lo que queremos y el sistema aprende las reglas por sí mismo.

### Ejemplo práctico: clasificación con Scikit-learn

Un ejemplo clásico de ML es clasificar flores iris según sus medidas:

````python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# Cargar datos y entrenar modelo
iris = load_iris()
modelo = RandomForestClassifier()
modelo.fit(iris.data, iris.target)

# Predecir una nueva flor
nueva_flor = [[5.1, 3.5, 1.4, 0.2]]
especie = iris.target_names[modelo.predict(nueva_flor)[0]]
print(f"Especie: {especie}")
````

El modelo aprende a distinguir 3 especies de iris basándose en 4 características. Recibe ejemplos etiquetados y **descubre los patrones** automáticamente.

## Deep Learning: aprendizaje profundo

El **Deep Learning** (DL) es una especialización del Machine Learning que utiliza **redes neuronales artificiales** con múltiples capas. Estas capas permiten al modelo aprender representaciones cada vez más abstractas de los datos.

La diferencia clave respecto al ML tradicional está en cómo se extraen las características:

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart TB
    subgraph ml["ML Tradicional"]
        direction LR

        M2 --> M3["Algoritmo"]
        M3 --> M4["Clasificación"]
    end

    subgraph dl["Deep Learning"]
        direction LR
        D1["Imagen<br/>(píxeles)"] --> D2["Red Neuronal<br/>aprende características<br/>automáticamente"]
        D2 --> D3["Clasificación"]
    end

    style ml fill:#ffebee,stroke:#c62828
    style dl fill:#e8f5e9,stroke:#2e7d32
```

Para reconocer gatos en imágenes, el ML tradicional requeriría que alguien definiera manualmente características como bordes, texturas y formas. Una red neuronal profunda recibe los píxeles directamente y **aprende por sí misma** qué patrones visuales identifican a un gato.

### Ejemplo práctico: red neuronal con PyTorch

Para construir una red neuronal que reconozca patrones complejos, podemos usar **PyTorch**:

````python
import torch
import torch.nn as nn

# Definir una red neuronal simple
class RedNeuronal(nn.Module):
    def __init__(self):
        super().__init__()
        self.capas = nn.Sequential(
            nn.Linear(784, 128), # Capa de entrada (ej: imagen 28x28)
            nn.ReLU(),           # Función de activación
            nn.Linear(128, 10)   # Capa de salida (10 clases)
        )

    def forward(self, x):
        return self.capas(x)

modelo = RedNeuronal()
print(modelo)
````

En este caso, definimos la **arquitectura** de la red (sus capas y neuronas). A diferencia del ML tradicional, aquí no le decimos al modelo qué buscar, sino que la red aprenderá a extraer sus propias características durante el entrenamiento.

> El Deep Learning elimina la necesidad de ingeniería manual de características, permitiendo abordar problemas donde definir reglas explícitas sería imposible.

## IA Generativa: de analizar a crear

La IA Generativa representa una especialización del Deep Learning con un objetivo diferente. Mientras que los sistemas de ML y DL se enfocan en **analizar y reconocer** patrones (IA discriminativa), la IA Generativa se centra en **crear contenido nuevo** (IA generativa).

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart LR
    subgraph tradicional["IA Discriminativa (ML/DL)"]
        direction TB
        T1["Imagen de gato"] --> T2["Modelo"]
        T2 --> T3["Etiqueta: 'Es un gato'"]
    end

    subgraph generativa["IA Generativa"]
        direction TB
        G1["'Un gato naranja<br/>durmiendo al sol'"] --> G2["Modelo<br/>Generativo"]
        G2 --> G3["Imagen NUEVA<br/>de un gato"]
    end

    tradicional -.->|"analiza"| A1["¿Qué ES esto?"]
    generativa -.->|"crea"| A2["¿Qué PODRÍA existir?"]

    style tradicional fill:#e3f2fd,stroke:#1976d2
    style generativa fill:#f3e5f5,stroke:#7b1fa2
```

| Característica | ML / DL tradicional | IA Generativa |
|----------------|---------------------|---------------|
| **Objetivo** | Analizar o clasificar | Crear contenido nuevo |
| **Entrada** | Datos a analizar | Instrucciones o prompts |
| **Salida** | Etiqueta, valor o categoría | Texto, imagen, audio, código |
| **Pregunta** | *¿Qué es esto?* | *¿Qué podrías crear?* |

### Cómo funciona la generación

Los modelos generativos aprenden la **distribución estadística** de los datos de entrenamiento. Durante la generación, no copian ejemplos existentes, sino que sintetizan contenido nuevo combinando los patrones aprendidos:

```mermaid
%%{init: {'theme': 'default'}}%%
flowchart TB
    subgraph entrenamiento["Fase de Entrenamiento"]
        E1["Millones de textos"] --> E2["Modelo aprende<br/>patrones del lenguaje"]
        E2 --> E3["Pesos del modelo<br/>(conocimiento comprimido)"]
    end

    subgraph generacion["Fase de Generación"]
        G1["Prompt del usuario"] --> G2["Modelo calcula<br/>siguiente token probable"]
        G2 --> G3["Token generado"]
        G3 -->|"se añade al contexto"| G2
        G3 --> G4["Respuesta completa"]
    end

    E3 -.->|"se usa"| G2

    style entrenamiento fill:#e8f5e9,stroke:#2e7d32
    style generacion fill:#fff3e0,stroke:#ef6c00
```

> La IA Generativa no busca en una base de datos: **calcula** qué contenido es más probable dado el contexto proporcionado.

### Ejemplo práctico: IA Generativa con LangChain

LangChain es el framework estándar para construir aplicaciones con modelos de lenguaje. En este ejemplo usamos **Ollama**, que permite ejecutar modelos localmente sin necesidad de API keys:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="deepseek-r1:1.5b")
respuesta = modelo.invoke("¿Por qué el cielo es azul?")
````

El modelo genera texto nuevo basándose en el contexto proporcionado. El parámetro `temperature` controla la creatividad (0=determinista, 1=creativo).
