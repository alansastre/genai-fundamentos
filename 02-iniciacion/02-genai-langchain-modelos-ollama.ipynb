## Qué significa el tamaño de un modelo

Al explorar la **biblioteca de modelos de Ollama** en [Ollama.com](https://www.ollama.com/search) encontrarás nombres como `gemma3:1b`, `llama3.2:3b`, `qwen3:8b` o `deepseek-r1:70b`. La letra "B" seguida de un número indica los **billones de parámetros** del modelo (en inglés, "billion" = mil millones).

Los **parámetros** son los valores numéricos que el modelo aprendió durante su entrenamiento. Puedes visualizarlos como millones de "interruptores" ajustados para reconocer patrones en el lenguaje y generar texto coherente. Cuantos más parámetros tiene un modelo, mayor capacidad tiene para capturar matices y relaciones complejas.

| Notación | Parámetros | Ejemplo de modelos |
|----------|------------|-------------------|
| **0.5b - 1b** | 500M - 1.000M | gemma3:1b, qwen3:0.6b |
| **1.5b - 4b** | 1.500M - 4.000M | deepseek-r1:1.5b, phi-4:3.8b |
| **7b - 9b** | 7.000M - 9.000M | llama3.1:8b, qwen3:8b, gemma3:4b |
| **12b - 14b** | 12.000M - 14.000M | gemma3:12b, qwen3:14b |
| **27b - 35b** | 27.000M - 35.000M | gemma3:27b, qwen3:32b |
| **70b - 72b** | 70.000M - 72.000M | llama3.3:70b, qwen2.5:72b |
| **235b+** | 235.000M+ | deepseek-r1:671b (versión completa) |

> Un modelo de 8B tiene 8.000 millones de parámetros. Esto no significa que sea "8 veces mejor" que uno de 1B, pero sí que puede manejar tareas más complejas y producir respuestas más elaboradas.

### Relación entre tamaño y capacidad

Los modelos más grandes generalmente ofrecen:

- **Mayor coherencia** en textos largos
- **Mejor comprensión** de instrucciones complejas
- **Mayor conocimiento** de hechos y conceptos
- **Mejor razonamiento** en problemas de múltiples pasos

Sin embargo, un modelo pequeño bien entrenado puede superar a uno grande mal entrenado. DeepSeek-R1:1.5b, por ejemplo, demuestra capacidades de razonamiento superiores a modelos de tamaño similar gracias a su entrenamiento especializado.

## Requisitos de hardware para ejecutar modelos

La pregunta más importante antes de descargar un modelo es: **¿puede mi ordenador ejecutarlo?** La respuesta depende principalmente de la **memoria disponible**, ya sea en la GPU (VRAM) o en la RAM del sistema.

### Regla general de memoria

Como referencia práctica, necesitas aproximadamente **la mitad del tamaño en parámetros en GB de memoria**. Esta regla asume modelos con cuantización Q4 (compresión estándar que reduce el tamaño sin perder demasiada calidad):

| Tamaño del modelo | Memoria mínima recomendada |
|-------------------|---------------------------|
| **1b - 3b** | 2 - 4 GB |
| **7b - 8b** | 4 - 6 GB |
| **12b - 14b** | 8 - 10 GB |
| **27b - 32b** | 16 - 20 GB |
| **70b** | 32 - 40 GB |

> Si tienes una GPU con 8 GB de VRAM, puedes ejecutar cómodamente modelos de hasta 8B-12B parámetros. Si solo tienes 4 GB, limítate a modelos de 3B-7B.

### GPU vs CPU: qué ocurre en cada caso

**Si tienes GPU compatible** (NVIDIA, AMD o Apple Silicon), Ollama la utilizará automáticamente. El modelo se carga en la memoria de la GPU (VRAM) y la inferencia es rápida:

- GPUs modernas: **40-80+ tokens por segundo**
- Respuestas casi instantáneas para preguntas simples
- Experiencia fluida e interactiva

**Si NO tienes GPU** (o tu GPU no tiene suficiente VRAM), Ollama ejecutará el modelo en la **CPU**, usando la RAM del sistema:

- CPUs modernas: **5-15 tokens por segundo**
- Las respuestas tardan varios segundos o minutos
- Funciona, pero la experiencia es más lenta

| Hardware | Modelo 8B | Modelo 70B |
|----------|-----------|------------|
| **GPU RTX 3060 (12GB)** | ~50 tok/s | No cabe en VRAM |
| **GPU RTX 4090 (24GB)** | ~80 tok/s | ~15 tok/s (parcial) |
| **CPU Intel i7 (32GB RAM)** | ~8 tok/s | ~2 tok/s |
| **Apple M2 (16GB unificada)** | ~35 tok/s | No cabe |

> Si no tienes GPU dedicada, no te preocupes. Para aprender y experimentar, los modelos pequeños (1B-8B) funcionan perfectamente en CPU. La velocidad es menor pero suficiente para pruebas.

Los modelos pequeños funcionan en la mayoría de equipos:

| Tu hardware | Modelo recomendado | Por qué |
|-------------|-------------------|---------|
| **Portátil básico (8GB RAM)** | gemma3:1b, deepseek-r1:1.5b | Consumen poca memoria, respuestas en segundos |
| **Ordenador medio (16GB RAM)** | gemma3:4b, llama3.2:3b | Mejor calidad manteniendo velocidad aceptable |
| **GPU 6-8GB VRAM** | qwen3:8b, llama3.1:8b | Aprovecha la GPU para respuestas rápidas |
| **GPU 12GB+ VRAM** | gemma3:12b, qwen3:14b | Modelos de alta calidad con contexto amplio |

> Empieza siempre por el modelo más pequeño de una familia (ej: gemma3:1b). Si funciona bien y quieres más calidad, prueba el siguiente tamaño (gemma3:4b, gemma3:12b). Así encontrarás el equilibrio óptimo para tu hardware.

## Descarga de modelos en Ollama

Ollama gestiona los modelos de lenguaje de forma similar a como Docker gestiona contenedores. Antes de poder utilizar un modelo, es necesario **descargarlo** al sistema local. Esta operación solo se realiza una vez por modelo.

Para descargar un modelo, se utiliza el comando `ollama pull` seguido del nombre y la etiqueta del modelo:

```bash .noeval
ollama pull gemma3:1b
```

Este comando descarga el modelo **Gemma 3** en su variante de 1 billón de parámetros. Los modelos pequeños como este son ideales para desarrollo y pruebas, ya que consumen menos recursos y responden más rápido.

> Los modelos se almacenan en `~/.ollama/models` en Linux/macOS y en `%LOCALAPPDATA%\ollama\models` en Windows. Puedes verificar el espacio disponible antes de descargar modelos grandes.

Para ver todos los modelos descargados en tu sistema:

```bash .noeval
ollama list
```

```bash .noeval
ollama pull deepseek-r1:1.5b
```

> DeepSeek-R1 es ideal para aprender sobre modelos de razonamiento sin necesitar hardware potente. Con solo 1.5 billones de parámetros, puede ejecutarse incluso en laptops básicas.

## Instanciar ChatOllama

La clase **ChatOllama** del paquete `langchain-ollama` es el punto de entrada para interactuar con modelos de Ollama desde Python. Esta clase implementa la interfaz estándar de modelos de chat de LangChain.

````python
from langchain_ollama import ChatOllama

# Instanciar el modelo
modelo = ChatOllama(model="gemma3:1b")
````

El parámetro `model` especifica qué modelo de Ollama utilizar. Debe coincidir exactamente con el nombre mostrado por `ollama list`.

### Verificar la conexión con el modelo

Antes de realizar inferencias complejas, es recomendable verificar que el modelo está correctamente configurado:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

# Verificar que el modelo responde
respuesta = modelo.invoke("Hola")
print(respuesta.content)
````

Si el modelo no está descargado o el servidor Ollama no está activo, este código lanzará una excepción con información sobre el problema.

## Invocar Gemma e interpretar la respuesta

El método `invoke` es la forma más directa de obtener una respuesta del modelo. Acepta una cadena de texto como entrada y devuelve un objeto **AIMessage** con la respuesta generada.

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")

# Invocar con texto simple
respuesta = modelo.invoke("¿Cuál es la capital de Francia?")

# El objeto respuesta es de tipo AIMessage
print(type(respuesta))
# <class 'langchain_core.messages.ai.AIMessage'>
````

El objeto **AIMessage** encapsula toda la información devuelta por el modelo. El atributo más importante es `content`, que contiene el texto generado:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")
respuesta = modelo.invoke("Explica qué es Python en una frase")

# Acceder al contenido de la respuesta
print(respuesta.content)
````

La salida será algo similar a:

```
Python es un lenguaje de programación interpretado, de alto nivel y propósito general, conocido por su sintaxis clara y legible.
```

> El atributo `content` siempre devuelve una cadena de texto con la respuesta del modelo. Este es el dato principal que utilizarás en la mayoría de aplicaciones.

### Metadatos de la respuesta

El objeto AIMessage también incluye información adicional sobre el proceso de generación en el atributo `response_metadata`:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="gemma3:1b")
respuesta = modelo.invoke("¿Qué es una API?")

# Metadatos de la generación
print(f"Modelo usado: {respuesta.response_metadata.get('model')}")
print(f"Razón de parada: {respuesta.response_metadata.get('done_reason')}")

# Información de tokens
print(f"Tokens de entrada: {respuesta.usage_metadata.get('input_tokens')}")
print(f"Tokens de salida: {respuesta.usage_metadata.get('output_tokens')}")
print(f"Total tokens: {respuesta.usage_metadata.get('total_tokens')}")
````

El campo `done_reason` indica por qué el modelo dejó de generar texto. El valor `stop` significa que el modelo completó su respuesta de forma natural.

## Invocar DeepSeek-R1 e interpretar el razonamiento

El modelo **DeepSeek-R1** tiene una característica distintiva: expone su proceso de razonamiento completo (chain-of-thought). Esto significa que puedes ver cómo el modelo "piensa" antes de dar su respuesta final.

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="deepseek-r1:1.5b")

respuesta = modelo.invoke("¿Cuánto es 15 multiplicado por 7?")
print(respuesta.content)
````

A diferencia de gemma3, la respuesta de deepseek-r1 incluye el proceso de razonamiento:

```
<think>
Para calcular 15 × 7, puedo descomponerlo:
15 × 7 = (10 × 7) + (5 × 7)
       = 70 + 35
       = 105
</think>

La respuesta es 105.
```

> El acceso al chain-of-thought de DeepSeek-R1 permite depurar y entender mejor cómo el modelo llega a sus conclusiones, lo cual es especialmente útil en tareas de razonamiento complejo.

### Comparar respuestas entre modelos

Una forma práctica de entender las diferencias entre modelos es hacer la misma pregunta a ambos:

````python
from langchain_ollama import ChatOllama

# Instanciar ambos modelos
gemma = ChatOllama(model="gemma3:1b")
deepseek = ChatOllama(model="deepseek-r1:1.5b")

pregunta = "Si tengo 3 manzanas y me dan el doble, ¿cuántas tengo?"

# Respuesta de Gemma
print("=== Gemma 3 ===")
respuesta_gemma = gemma.invoke(pregunta)
print(respuesta_gemma.content)

# Respuesta de DeepSeek-R1 con razonamiento
print("\n=== DeepSeek-R1 ===")
respuesta_deepseek = deepseek.invoke(pregunta)
print(respuesta_deepseek.content)
````

La respuesta de **Gemma** será directa:

```
Tendrías 6 manzanas.
```

La respuesta de **DeepSeek-R1** mostrará el razonamiento:

```
<think>
Veamos paso a paso:
1. Tengo inicialmente 3 manzanas
2. Me dan el doble de lo que tengo
3. El doble de 3 es: 3 × 2 = 6 manzanas
4. Total: 3 + 6 = 9 manzanas
</think>

Tendrías 9 manzanas en total.
```

Este ejemplo ilustra cómo diferentes modelos pueden interpretar la misma pregunta de formas distintas. DeepSeek-R1 tiende a desglosar el problema y mostrar cada paso dentro de las etiquetas `<think>`, mientras que modelos más pequeños dan respuestas más concisas.

> El modelo DeepSeek-R1 es especialmente útil cuando necesitas verificar el razonamiento detrás de una respuesta o cuando trabajas con problemas que requieren múltiples pasos de lógica.

### Estructura del AIMessage con DeepSeek-R1

El objeto AIMessage devuelto por DeepSeek-R1 tiene la misma estructura que el de cualquier otro modelo, pero el contenido incluye tanto el razonamiento (entre etiquetas `<think>`) como la respuesta final:

````python
from langchain_ollama import ChatOllama

modelo = ChatOllama(model="deepseek-r1:1.5b")
respuesta = modelo.invoke("¿Por qué el cielo es azul?")

# El content incluye todo el proceso de pensamiento
print(f"Longitud de la respuesta: {len(respuesta.content)} caracteres")
print(f"Tokens generados: {respuesta.usage_metadata.get('output_tokens')}")
````

Las respuestas de DeepSeek-R1 suelen ser más largas debido al razonamiento incluido, lo que se refleja en un mayor número de tokens de salida. Esto es importante tenerlo en cuenta al planificar el uso de recursos.

### Ventajas de DeepSeek-R1 para aprendizaje

- **Tamaño accesible**: con solo 1 GB, cualquier estudiante puede descargarlo
- **Razonamiento visible**: permite aprender cómo los modelos descomponen problemas
- **Código abierto**: puedes ejecutarlo localmente sin costes de API
- **Balance**: combina capacidades de razonamiento con requisitos mínimos de hardware
